### 2.2.2 性能指標を選択する
次に性能指標を検討する．回帰問題の典型的な性能指標は以下の平均二乗誤差（Root Mean Square Error: RMSE）である．

$$
    RMSE({\bf X}, h) = \sqrt{\frac{1}{m} \sum_{i=0}^m \left(h\left({\bf x}^{(i)}\right) - y^{(i)}\right)^2}
$$

平均絶対誤差（mean absolute error : MAE）を使う場合もある．MAE は平均絶対偏差（mean absolute deviation）とも呼ばれる．

$$
    MAE({\bf X}, h) = \frac{1}{m} \sum_{i=0}^m \left| h\left({\bf x}^{(i)}\right) - y^{(i)} \right|
$$

上記はどちらも2つのベクトルの距離を図る指標で，一般にノルム（norm）と呼ばれる．他にも様々なノルムがある．
- RMSE はユークリッドノルム（Euclidian norm）に対応し，$l_2$ノルムとも呼ばれ，$\Arrowvert \cdot \Arrowvert_2$ や $\Arrowvert \cdot \Arrowvert$ と記す
- MAE は $l_1$ ノルムに対応し，マンハッタンノルムとも呼ぶ
- 一般の $l_k$ ノルムは $\Arrowvert {\bf v} \Arrowvert_k = l_0\sqrt[n]{|v_0|^k+|v_1|^k+ \cdots +|v_n|^k}$ で定義される
（$l_0：ベクトルの非ゼロ要素の数，$$l_{\infty}$：ベクトルの絶対師の最大）
- のツムの数字が大きいほど，大きな値を重視し小さい値は無視する傾向が強くなる

```python
# データセットをダウンロードして CSV に展開する
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
    
fetch_housing_data()
```

### 2.3.4 テストセットを作る
これまでデータをざっと見ただけだが，あまり詳しく見てしまうと思いがけないパターンを見つけてしまい，特定のモデルを選ぶように誘導されてしまうかもしれない．このように，データを盗み見することで推定が楽観的になりすぎることを，**データスヌーピングバイアス（data snooping bias）**と呼ぶ．これを防ぐため，この段階でデータの一部をテストセットとして切り分けて封印する．  
テストセットの作成は，無作為に一部のインスタンス（一般的にはデータセットの20%）を取り出す．

```python
def split_train_test(data, test_ratio=0.2):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

train_set, test_set = split_train_test(housing)
print(len(train_set), "train +", len(test_set), "test")
```

これでテストセットの作成はできるのだが，この方法は完璧ではない．
プログラムを実行するたびに異なるテストセットが作成されてしまい，結局データセット全体を見てしまうことになりかねない．
これを防ぐために，最初の実行で使ったテストセットを残しておいて，以降の実行時にそれを読み込むという方法がある．
また，numpyの乱数生成器のseed（たとえばnp.random.seed(42)）を設定しておくこともある．  
しかし，これらの方法では，データセットが更新された瞬間に破綻してしまう．
そのため，以下のように各インスタンスに識別子をつけておき，そのインスタンスがテストセットに含まれるものかどうかを判断する方法をとる．
こうすれば，データセットが更新されたときもセストセットはイッテに保たれる．

```python
def test_set_check(identifier, test_ratio, hash):
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio

def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

しかし，住宅価格データセットには識別子の列はない．
このような場合，もっとも単純な方法は行番号をIDにすることである．

```python
# ID 列を追加する
housing_with_id = housing.reset_index()
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
```

行番号を一意な識別子として使う場合には，新データがデータセットの末尾に追加されるようにして，行が削除されないようにしなければならない．
これ以外の方法としては，区域の緯度と経度を組み合わせて識別子とすることもできる．

```python
# 緯度と経度を組み合わせて識別子とする
housing_with_id["id"] = housing["longitude"] * 1000 + housing["latitude"]
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
```

scikit-learnには上記のsplit_train_testとほぼ同等の機能を持った関数が実装されている．
この関数には，乱数生成器のseedを設定するrandom_state引数を取り，複数のデータセットに同じ行番号を与え，同じインデックスでデータセットを分割する機能が盛り込まれている．

```python
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
```

ここまで見てきた分割方法は，純粋に無作為なサンプリング方法である．
これはデータセットが十分に大規模なら問題ないが，そうでない場合にはサンプリングバイアスを持ち込むおそれがある．  
例えば，調査会社が1000人を対象にし調査を行うとき，アメリカの人口比，女性51.3%，男性48.7%に従って，513人の女性と487人の男性に対して調査を行う．このような調査方法を，**層化抽出法（stratified sampling）**と呼ぶ．
こうして集められたデータセットに対して，純粋に無作為なサンプリングを行うと，一定の確率で歪んだ検証セットをサンプリングしてしまう．  

さて，収入の中央値が住宅価格の中央値を予測するのに非常に重要な属性だとわかったとする．
テストセットはデータセット全体の様々な収入カテゴリを代表したものにしたい．
まずは，収入カテゴリという属性を新たに作る．
収入の中央値のヒストグラムをよく見ると，収入の中央値の大半は2万〜5万ドルに集中しているが，一部の値は6万ドルを大きく越えている．
データセットの各層に十分な数のインスタンスがあることが重要で，そうでなければ層を重視することがバイアスになってしまう．
つまり，層の数が多くなりすぎないようにしながら，各層は十分に大きくなければならない．  

次のコードでは，収入の中央値を1.5で割り，端数切り上げの上，5以上のカテゴリを全て5にまとめている．

```python
# 収入カテゴリを新しく作る
housing["income_cat"] = np.ceil(housing["median_income"] / 1.5)
housing["income_cat"].where(housing["income_cat"] < 5, 5.0, inplace=True)

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
housing["income_cat"].hist()
```

```python
strat_test_set["income_cat"].value_counts() / len(strat_test_set)
```

上記のように収入カテゴリごとの割合を見ることができる．
また，同様なコードでデートセット全体の収入カテゴリの割合も調べられる．

```python
def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

compare_props = pd.DataFrame({
    "Overall": income_cat_proportions(housing),
    "Stratified": income_cat_proportions(strat_test_set),
    "Random": income_cat_proportions(test_set),
}).sort_index()
compare_props["Rand. %error"] = 100 * compare_props["Random"] / compare_props["Overall"] - 100
compare_props["Strat. %error"] = 100 * compare_props["Stratified"] / compare_props["Overall"] - 100

compare_props
```

最後にデータを元の状態に戻しておく．

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
```

## 2.5 機械学習アルゴリズムに渡せるようにデータを準備する
前節までの内容を踏まえて，機械学習アルゴリズムに渡すデータの準備を行う．次のような理由から，準備作業をまとめた関数を作成しておきたい．

- 関数を用意しておけば，どんなデータセットに対しても簡単に変換操作を再現できる
- 将来のプロジェクトで再利用可能な変換関数のライブラリが整備されていく
- 本番システムに組み込むことで，新しいデータをアルゴリズムに与える前に変換できる
- 様々な変換方式を試し，どの組み合わせが最もよく機能するか試せる

最初に，クリーンな訓練セットに戻した上で予測子とラベルを分けておく．

```python
housing = strat_train_set.drop("median_house_value", axis=1)
housing_labels = strat_train_set["median_house_value"].copy()
```

### 2.5.1 データをクリーニングする
殆どの機械学習アルゴリズムでは欠損特徴量の処理ができないため，それに対応するための関数を作っておく．
今使っているデータセットでは，total_bedrooms 属性に欠損値があるため，以下の3通りのいずれかで対処する．

1. 対応する区域を取り除く
1. 属性全体を取り除く
1. 何らかの値を設定する（0，平均値，中央値など）

これらはそれぞれ，以下のように実装できる．

```python
# 1. 対応する区域の除外
housing.dropna(subset=["total_bedrooms"])
# 2. 属性全体の削除
housing.drop("total_bedrooms", axis=1)
# 3. 中央値で補完
median = housing["total_bedrooms"].median()
housing["total_bedrooms"].fillna(median, inplace=True)
```

あとから参照できるように3つ目の方法では中央値を保存しておくようにする．  

このように，マニュアルで対処する以外に，scikit-learn には SimpleImputer という欠損値をよしなに処理してくれるクラスがあるのでそれを使っても良い．

```python
imputer = SimpleImputer(strategy="median")
```

### 2.5.3 カスタム変換
次に示すスクリプトは寝室数と部屋数を結合した属性を追加するための変換器クラスである．
scikit-learn では fit()，transform()，fit_transform()，の3つのメソッドを実装すれば，独自変換器を追加できる．
メソッドのうちあとの2つは，基底クラスに TransformerMixin() を基底クラスに追加すれば手に入る．
また，基底クラスに BaseEstimator を追加すれば，自動的にハイパーパラメータをチューニングに使う get_params() と set_params() のメソッドが使えるようになる．

```python
rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

        
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```

### 2.5.4 特徴量スケーリング
**特徴量スケーリング**（feature scaling）は事前処理の中でも特に重要である．
基本的に機械学習アルゴリズムは数値属性のスケールが大きく異なると性能を発揮できない．
全ての数値属性のスケールを統一するため，よく使われる手法としては**最小最大スケーリング**（min-max scaling）と**標準化**（standarization）がある．  
最小最大スケーリングは一般的には**正規化**（normalization）と呼ばれる．
この方法では，単純に全ての値が0から1に収まるようにスケーリングするだけである．属性値から最小値を引き，最小値と最大値の差で割れば良い．
scikit-learn では MinMaxScaler でこの変換が可能である．また，feature_range 引数で値の範囲を調整できる．  
標準化は属性値から平均値を引き，さらに分散で割って得られる分布の分散が単位分散となるようにする．この方法では値の上限や下限がないため，それが問題になることもあるが，外れ値の影響を抑えることができるという利点もある．
scikit-learn には StandardScaler が用意されている．

### 2.5.5 変換パイプライン
データ変換では必要なステップがいくつもあり，それを正しい順序で実行する必要がある．
scikit-learn では Pipeline クラスを使って，変換シーケンスを実行しやすくできる．

```python
num_pipline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler())
])

housing_num_tr = num_pipline.fit_transform(housing_num)
```

Pipeline のコンストラクタは引数として，変換ステップのシーケンスを定義する名前/推定器のペアのリストを取る．
パイプラインの fit() メソッドを呼び出すと，全ての変換器の fit_transform() が逐次的に呼び出される．
このとき，引数として前段の変換器の出力が渡される．
最後の推定器では fit() メソッドが呼び出される．
パイプラインは最後の推定器と同じメソッドを外部から呼び出せるようにしている．  

ここで，Pandas の数値以外のカラムを含んでいる DataFrame を与えることができると嬉しい．
scikit-learn には，DataFrame を処理する機能はないため，簡単なカスタム変換器を作る

```python
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X[self.attribute_names].values
```

これによって，必要な属性だけを選択し，残りを破棄して DataFrame の結果を NumPy 配列に変換する．
最初に示したパイプラインを次のように書き換えることで， DataFrame を利用して数値のみを処理するパイプラインを作成できる．

```python
num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

num_pipeline = Pipeline([
    ('selector', DataFrameSelector(num_attribs)),
    ('imputer', SimpleImputer(strategy="median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('selector', DataFrameSelector(cat_attribs)),
    ('cat_encoder', OneHotEncoder(sparse=False))
])
```

これら2つのパイプラインを1つにまとめたい．
これを実現するために，FeatureUnion() メソッドが用意されている．
各変換器のtransform()メソッドが並列に実行され，それぞれの出力を待って結合した上で結果を返す．

```python
full_pipeline = FeatureUnion(transformer_list=[
    ("num_pipeline", num_pipeline),
    ("cat_pipeline", cat_pipeline)
])
```

```python
housing_prepared = full_pipeline.fit_transform(housing)
housing_prepared
```

### 2.6.2 交差検証を使ったより良い評価
モデルを評価する簡単な方法は，train_test_split() 関数で訓練セットを小さい訓練セットと検証セットに分割し，小さい方の訓練セットでモデルを訓練し，検証セットで評価するという方法である．  
もう一つの方法は**交差検証** （cross-validation）である．次に示すコードは**K分割交差検証**（K-fold cross-validation）と呼ばれる．
ここでは，訓練セットを10個の**フォールド**（fold）に分割し，1個のフォールドを評価用に残して，その他の9個のフォールドで訓練している．

```python
# 決定木による回帰
scores = cross_val_score(tree_reg, housing_prepared, housing_labels, scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)

def display_scores(scores):
    print("Scores: ", scores)
    print("Mean: ", scores.mean())
    print("Standard deviation: ", scores.std())
    
display_scores(tree_rmse_scores)
```

```python
# 線形回帰モデル
lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels, scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

このように決定木による回帰と線形回帰を交差検証で比較すると，決定木の方は過学習の度合いが大きく，線形回帰モデルよりもかえって性能が落ちてしまっている．
最後に，ランダムフォレストによる回帰を試してみる．
ランダムフォレストは特徴量の無作為なサブセットを使って複数の決定木を学習し，それらの予測の平均を取る．
このような学習手法を**アンサンブル学習**（ensemble learning）と呼ぶ．

```python
forest_reg = RandomForestRegressor()
forest_reg.fit(housing_prepared, housing_labels)
forest_scores = cross_val_score(forest_reg, housing_prepared, housing_labels, scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```

これでかなりスコアが改善している．しかし，まだ過学習しているように見える．
今後の目標としては，いくつかの機械学習アルゴリズムを試して，有望なモデルのリストを得ることになる．

### 2.7.1 グリッドサーチ
マニュアルで最適なハイパーパラメータを見つけるのは非常に手間がかかる．
そこで，scikit-learn の GridSearchCV を使って，指定されたハイパーパラメータの組み合わせを全て交差検証によって評価する．
次のコードはランダムフォレストのパラメータ値を探索するものである．

```python
param_grid = [
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]}
]

forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error')

grid_search.fit(housing_prepared, housing_labels)
```

```python
grid_search.best_params_
```

上のコードでは最初に param_grid の1つ目の辞書に含まれる12通りのハイパーパラメータの組み合わせを評価し，次に2つ目の辞書で指定されたパラメータの組み合わせ6通りを評価している．また，それぞれのモデルについて5回ずつ訓練を行っている．
これによって，最良の推定器を得ることができる．

```python
grid_search.best_estimator_
```

また，評価のスコアも次のように得られる．

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

### 2.7.2 ランダムサーチ
グリッドサーチは，前節のように少数のハイパーパラメータの組み合わせを探索する際には適しているが，**探索空間**（search space）が大きいときには不適である．
そのような場合には，RandomizedSearchCV を使う．
これは全ての組み合わせを試すのではなく，指定回数だけランダムにハイパーパラメータを設定し，それらの組み合わせを評価する．
このアプローチには次のようなメリットがある．

- 個々のハイパーパラメータについて多くの異なる値を試せる
- イテレーション回数の指定だけで済むので，計算資源の管理がしやすくなる
